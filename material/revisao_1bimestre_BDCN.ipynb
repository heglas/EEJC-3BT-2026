## 🧩 PARTE 1 — Modelagem de Dados e Diagrama ER

### 📖 Conceitos-chave

**Modelagem de Dados** é o processo de descrever como os dados serão organizados antes de criar o banco.  
Existem três níveis de modelagem:

| Nível | Descrição | Foco |
|-------|-----------|------|
| **Conceitual** | Visão de negócio, independente de tecnologia | Entidades e Relacionamentos |
| **Lógico** | Estrutura com atributos e chaves, independente de SGBD | Tabelas, PK, FK |
| **Físico** | Implementação real no SGBD escolhido | SQL `CREATE TABLE` |

---

### 🔑 Elementos do Diagrama ER

- **Entidade** → objeto do mundo real com existência independente (ex: `Aluno`, `Livro`, `Professor`)
- **Atributo** → característica de uma entidade (ex: `nome`, `data_nascimento`, `cpf`)
- **Relacionamento** → associação entre entidades (ex: `Aluno` *empréstimo* `Livro`)
- **Chave Primária (PK)** → identifica de forma única cada registro da tabela
- **Chave Estrangeira (FK)** → referencia a PK de outra tabela, criando o relacionamento
- **Cardinalidade** → define quantos registros de uma entidade se relacionam com outra (1:1, 1:N, N:M)

---

### 📐 Exemplo — Diagrama ER de uma Biblioteca

```
┌─────────────┐         ┌───────────────┐         ┌──────────────┐
│   ALUNO     │         │  EMPRÉSTIMO   │         │    LIVRO     │
│─────────────│  1   N  │───────────────│  N   1  │──────────────│
│ PK id_aluno │─────────│ PK id_empr.   │─────────│ PK id_livro  │
│ nome        │         │ FK id_aluno   │         │ titulo       │
│ email       │         │ FK id_livro   │         │ autor        │
│ data_nasc.  │         │ data_empr.    │         │ isbn         │
└─────────────┘         │ data_devol.   │         └──────────────┘
                        └───────────────┘
```

## 🛢️ PARTE 2 — SQL: SELECT, FROM e WHERE

### 📖 Estrutura básica de uma consulta SQL

```sql
SELECT coluna1, coluna2    -- O QUE mostrar
FROM nome_da_tabela        -- DE ONDE vem
WHERE condição;            -- FILTRO (opcional)
```

> 💡 **Lembre-se:** A cláusula `WHERE` sempre vem APÓS a especificação da tabela (`FROM`).

### 🔢 2.1 — Operadores de Comparação com Números

| Operador | Significado        | Exemplo SQL                        |
|----------|--------------------|------------------------------------|
| `=`      | Igual              | `WHERE idade = 16`                 |
| `<>`     | Diferente          | `WHERE idade <> 15`                |
| `>`      | Maior que          | `WHERE altura > 1.80`              |
| `>=`     | Maior ou igual     | `WHERE altura >= 1.80`             |
| `<`      | Menor que          | `WHERE altura < 1.70`              |
| `<=`     | Menor ou igual     | `WHERE idade <= 15`                |


````python
# Simulando a tabela 'alunos' em Python para praticar os conceitos de SQL
import pandas as pd

alunos = pd.DataFrame({
    'id':   [1,      2,       3,      4,        5,     6],
    'nome': ['João','Lucas','Gabi','Márcio','Ana','Jéssica'],
    'altura':[1.80, 1.67,   1.76,  1.94,   1.55,  1.63],
    'idade': [16,   15,     15,    15,     16,    16],
    'cidade':['São Paulo','Rio de Janeiro','São Paulo','Curitiba','São Roque','São Paulo'],
    'data_nascimento':['2007-02-25','2008-05-23','2008-03-12','2008-09-23','2007-10-05','2007-05-08']
})

print("📋 Tabela completa de alunos:")
print(alunos.to_string(index=False))
````

#### 🔍 Exemplos com números

**Exemplo 1:** Alunos com idade igual a 16
```sql
SELECT * FROM alunos WHERE idade = 16;
```


````python
# Exemplo 1: WHERE idade = 16
resultado = alunos[alunos['idade'] == 16]
print("Alunos com idade = 16:")
print(resultado[['nome','idade','cidade']].to_string(index=False))
````

**Exemplo 2:** Alunos com altura maior que 1.80
```sql
SELECT * FROM alunos WHERE altura > 1.80;
```


````python
# Exemplo 2: WHERE altura > 1.80
resultado = alunos[alunos['altura'] > 1.80]
print("Alunos com altura > 1.80:")
print(resultado[['nome','altura']].to_string(index=False))
````

### 📝 2.2 — WHERE com Textos e LIKE

> ⚠️ **Atenção:** Comparações de texto são **sensíveis a maiúsculas/minúsculas** e acentos!

| Operador | Uso | Exemplo |
|----------|-----|---------|
| `=`      | Texto exatamente igual | `WHERE cidade = 'São Paulo'` |
| `<>`     | Texto diferente | `WHERE cidade <> 'Curitiba'` |
| `LIKE`   | Busca parcial com coringa `%` | `WHERE cidade LIKE 'São%'` |

**Como funciona o `%` no LIKE:**
- `'São%'` → começa com "São"
- `'%Paulo'` → termina com "Paulo"
- `'%ao%'` → contém "ao" em qualquer posição

**Exemplos:**
```sql
-- Alunos de São Paulo (exato)
SELECT * FROM alunos WHERE cidade = 'São Paulo';

-- Cidades que começam com 'São'
SELECT * FROM alunos WHERE cidade LIKE 'São%';
```


````python
# WHERE com texto — igualdade exata
resultado = alunos[alunos['cidade'] == 'São Paulo']
print("Alunos de São Paulo (exato):")
print(resultado[['nome','cidade']].to_string(index=False))
````

````python
# WHERE com LIKE — cidades que começam com 'São'
resultado = alunos[alunos['cidade'].str.startswith('São')]
print("Cidades que começam com 'São' (equivalente ao LIKE 'São%'):")
print(resultado[['nome','cidade']].to_string(index=False))
````

### ✅ Atividade 3 — Consultas com texto

1. Escreva uma query SQL para buscar alunos que têm a letra **'a'** no nome (minúscula).
2. Escreva uma query para buscar alunos cuja cidade **termina** com 'Paulo'.


````python
# Atividade 3.1 — Alunos com letra 'a' no nome
# SQL: SELECT * FROM alunos WHERE nome LIKE '%a%';

resultado_3a = alunos[alunos['nome'].str.contains('a')]
print("Alunos com 'a' no nome:")
print(resultado_3a[['nome']].to_string(index=False))

print()

# Atividade 3.2 — Cidades que terminam com 'Paulo'
# SQL: SELECT * FROM alunos WHERE cidade LIKE '%Paulo';

resultado_3b = alunos[alunos['cidade'].str.endswith('Paulo')]
print("Cidades que terminam com 'Paulo':")
print(resultado_3b[['nome','cidade']].to_string(index=False))
````

### 📅 2.3 — WHERE com Datas

> 💡 **Formato padrão SQL:** `AAAA-MM-DD` (sempre entre aspas simples)  
> Datas são ordenadas **cronologicamente** — uma data "maior" é mais recente.

```sql
-- Alunos nascidos exatamente em 05/10/2007
SELECT * FROM alunos WHERE data_nascimento = '2007-10-05';

-- Alunos nascidos ANTES de 01/01/2008
SELECT * FROM alunos WHERE data_nascimento < '2008-01-01';
```


````python
# WHERE com datas — nascidos antes de 2008
alunos['data_nascimento'] = pd.to_datetime(alunos['data_nascimento'])

resultado = alunos[alunos['data_nascimento'] < '2008-01-01']
print("Alunos nascidos antes de 2008:")
print(resultado[['nome','data_nascimento']].to_string(index=False))
````

---
### 🔗 2.4 — Operadores Lógicos: AND e OR

Permitem **combinar duas ou mais condições** no WHERE.

| Operador | Regra | Exemplo |
|----------|-------|---------|
| `AND` | **Ambas** as condições devem ser verdade | `WHERE idade = 16 AND cidade = 'São Paulo'` |
| `OR`  | **Pelo menos uma** condição verdade | `WHERE cidade = 'Curitiba' OR cidade = 'São Roque'` |

**Tabela-verdade:**

| V1 | V2 | V1 AND V2 | V1 OR V2 |
|----|----|-----------|----------|
| V  | V  | **V**     | V        |
| V  | F  | F         | **V**    |
| F  | V  | F         | **V**    |
| F  | F  | F         | F        |

---

**Exemplo clássico:** Alunos nascidos no 2º semestre de 2007:
```sql
SELECT * FROM alunos
WHERE data_nascimento > '2007-06-30'
  AND data_nascimento < '2008-01-01';
```


````python
# AND — nascidos no 2º semestre de 2007
resultado = alunos[
    (alunos['data_nascimento'] > '2007-06-30') &
    (alunos['data_nascimento'] < '2008-01-01')
]
print("Alunos nascidos no 2º semestre de 2007:")
print(resultado[['nome','data_nascimento']].to_string(index=False))
````

````python
# OR — alunos de Curitiba OU São Roque
resultado = alunos[
    (alunos['cidade'] == 'Curitiba') |
    (alunos['cidade'] == 'São Roque')
]
print("Alunos de Curitiba ou São Roque:")
print(resultado[['nome','cidade']].to_string(index=False))
````

### ✅ Atividade 4 — Operadores Lógicos

1. Busque alunos que têm **mais de 1.70m** E são de **São Paulo**.
2. Busque alunos com **idade = 16** OU **altura > 1.90**.
3. (**Desafio**) Busque alunos nascidos entre **01/09/2007** e **01/01/2008**.


````python
# Atividade 4.1 — altura > 1.70 AND cidade = 'São Paulo'
# SQL: SELECT * FROM alunos WHERE altura > 1.70 AND cidade = 'São Paulo';

r1 = alunos[(alunos['altura'] > 1.70) & (alunos['cidade'] == 'São Paulo')]
print("Altura > 1.70 E cidade São Paulo:")
print(r1[['nome','altura','cidade']].to_string(index=False))

print()

# Atividade 4.2 — idade = 16 OR altura > 1.90
# SQL: SELECT * FROM alunos WHERE idade = 16 OR altura > 1.90;

r2 = alunos[(alunos['idade'] == 16) | (alunos['altura'] > 1.90)]
print("Idade = 16 OU altura > 1.90:")
print(r2[['nome','idade','altura']].to_string(index=False))

print()

# Atividade 4.3 — DESAFIO: nascidos entre set/2007 e jan/2008
# SQL: SELECT * FROM alunos WHERE data_nascimento >= '2007-09-01' AND data_nascimento < '2008-01-01';

r3 = alunos[
    (alunos['data_nascimento'] >= '2007-09-01') &
    (alunos['data_nascimento'] < '2008-01-01')
]
print("Nascidos entre set/2007 e jan/2008 (desafio):")
print(r3[['nome','data_nascimento']].to_string(index=False))
````

### Exercício 4.3 — Overfitting vs Underfitting
Execute e analise o código abaixo.

```python
from sklearn.tree import DecisionTreeClassifier

profundidades = range(1, 16)
acc_treino, acc_teste = [], []

for d in profundidades:
    dt = DecisionTreeClassifier(max_depth=d, random_state=42)
    dt.fit(Xf_train, yf_train)
    acc_treino.append(accuracy_score(yf_train, dt.predict(Xf_train)))
    acc_teste.append(accuracy_score(yf_test,  dt.predict(Xf_test)))

plt.figure(figsize=(9, 5))
plt.plot(profundidades, acc_treino, 'o-', label='Treino',  color='steelblue')
plt.plot(profundidades, acc_teste,  's-', label='Teste',   color='orange')
plt.axvline(x=profundidades[np.argmax(acc_teste)],
            color='red', linestyle='--', label='Melhor profundidade')
plt.xlabel('Profundidade Máxima da Árvore')
plt.ylabel('Acurácia')
plt.title('Overfitting vs Underfitting — Decision Tree')
plt.legend()
plt.grid(True)
plt.show()

melhor = profundidades[np.argmax(acc_teste)]
print(f'Melhor profundidade: {melhor} | Acurácia no teste: {max(acc_teste):.4f}')
```

## 🐼 Parte 2 — Fundamentos do Pandas

### 2.1 Criando Series e DataFrames


```python
# --- Series ---
serie_notas = pd.Series([8.5, 7.0, 9.2, 6.8, 10.0],
                        index=['Acácio', 'Artur', 'Ayssa', 'Brian', 'Bruna'])
print('Serie de notas da turma 3BT:')
print(serie_notas)
```

```python
# --- DataFrame ---
dados_alunos = {
    'Nome': ['Acácio', 'Artur', 'Ayssa', 'Brian', 'Bruna', 'Felipe', 'Gabriel A.', 'Gabriela'],
    'Nota_1B': [8.5, 7.0, 9.2, 6.8, 10.0, 7.5, 8.8, 9.0],
    'Faltas': [2, 0, 1, 4, 0, 3, 1, 0],
    'Situação': ['Aprovado', 'Aprovado', 'Aprovado', 'Recuperação', 'Aprovado',
                 'Aprovado', 'Aprovado', 'Aprovado']
}

df = pd.DataFrame(dados_alunos)
print('DataFrame da Turma 3BT — EEJC:')
df
```

### 2.2 Informações básicas do DataFrame


```python
# Explorando o DataFrame
print('Shape:', df.shape)
print('\nColunas:', df.columns.tolist())
print('\nTipos de dados:')
print(df.dtypes)
print('\nEstatísticas descritivas:')
df.describe()
```

---
## 🔍 Parte 3 — Acesso e Seleção de Dados

### 3.1 Selecionando colunas


```python
# Selecionando uma única coluna
print(df['Nome'])

# Selecionando múltiplas colunas
print(df[['Nome', 'Nota_1B']])
```

### 3.2 Filtragem condicional


```python
# Alunos em recuperação
recuperacao = df[df['Situação'] == 'Recuperação']
print('Alunos em recuperação:')
print(recuperacao)

# Alunos com nota >= 8.0
destaque = df[df['Nota_1B'] >= 8.0]
print('\nAlunos com nota >= 8.0:')
print(destaque[['Nome', 'Nota_1B']])
```

### 3.3 Usando `.loc` e `.iloc`


```python
# .loc — por rótulo
print('Linha 0 (loc):')
print(df.loc[0])

# .iloc — por posição numérica
print('\nPrimeiros 3 alunos (iloc):')
print(df.iloc[0:3])

# Célula específica
print('\nNota do Artur (loc):', df.loc[1, 'Nota_1B'])
```

---
## 🔢 Parte 4 — NumPy: Operações com Arrays


```python
# Criando arrays NumPy
notas = np.array([8.5, 7.0, 9.2, 6.8, 10.0, 7.5, 8.8, 9.0])

print('Array de notas:', notas)
print('Média:', np.mean(notas))
print('Mediana:', np.median(notas))
print('Desvio Padrão:', np.std(notas).round(2))
print('Nota máxima:', np.max(notas))
print('Nota mínima:', np.min(notas))
```

```python
# Operações vetorizadas
# Simulando bônus de 0.5 ponto para quem tem faltas == 0
faltas = np.array([2, 0, 1, 4, 0, 3, 1, 0])
bonus = np.where(faltas == 0, 0.5, 0)
notas_finais = np.clip(notas + bonus, 0, 10)  # máximo 10
print('Notas com bônus de frequência:', notas_finais)
```

## 📈 Parte 5 — Visualização Gráfica com Pandas

O método `.plot()` do Pandas usa o parâmetro `kind` para definir o tipo de gráfico.  
A tabela abaixo resume os principais tipos estudados:

| `kind=`      | Tipo de Gráfico      | Melhor usado para... |
|--------------|----------------------|----------------------|
| `'line'`     | Linhas               | Tendências ao longo do tempo |
| `'bar'`      | Barras verticais     | Comparação de categorias |
| `'barh'`     | Barras horizontais   | Categorias com nomes longos |
| `'hist'`     | Histograma           | Distribuição de frequência |
| `'box'`      | Boxplot              | Dispersão e outliers |
| `'pie'`      | Pizza / Setores      | Proporções de um todo |
| `'scatter'`  | Dispersão            | Correlação entre variáveis |
| `'area'`     | Área                 | Volume acumulado no tempo |
| `'kde'`      | Densidade de kernel  | Distribuição suavizada |


### 5.1 Gráfico de Barras — Notas por Aluno


```python
df_plot = df[['Nome', 'Nota_1B']].set_index('Nome')
ax = df_plot.plot(kind='bar', color='steelblue', figsize=(10, 5), legend=False)
ax.set_title('Notas do 1º Bimestre — Turma 3BT EEJC', fontsize=14)
ax.set_xlabel('Aluno')
ax.set_ylabel('Nota')
ax.axhline(y=5.0, color='red', linestyle='--', label='Média mínima (5.0)')
ax.legend()
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 5.2 Histograma — Distribuição das Notas


```python
ax = df_plot.plot(kind='hist', bins=5, edgecolor='black', color='coral', figsize=(8, 4), legend=False)
ax.set_title('Distribuição das Notas — Turma 3BT EEJC', fontsize=14)
ax.set_xlabel('Nota')
ax.set_ylabel('Frequência')
plt.tight_layout()
plt.show()
```

### 5.3 Boxplot — Variação das Notas


```python
ax = df[['Nota_1B', 'Faltas']].plot(kind='box', figsize=(7, 5))
ax.set_title('Boxplot — Notas e Faltas (Turma 3BT)', fontsize=13)
plt.tight_layout()
plt.show()
```

### 5.4 Gráfico de Pizza — Situação dos Alunos


```python
situacao_counts = df['Situação'].value_counts()
ax = situacao_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90,
                          colors=['#4CAF50', '#FF5722'], figsize=(6, 6))
ax.set_title('Situação dos Alunos — Turma 3BT EEJC', fontsize=13)
ax.set_ylabel('')
plt.tight_layout()
plt.show()
```

### 5.5 Gráfico de Dispersão — Notas × Faltas


```python
ax = df.plot(kind='scatter', x='Faltas', y='Nota_1B',
            color='purple', figsize=(7, 5), s=80)
ax.set_title('Dispersão: Faltas × Nota — Turma 3BT EEJC', fontsize=13)
ax.set_xlabel('Número de Faltas')
ax.set_ylabel('Nota do 1º Bimestre')
plt.tight_layout()
plt.show()
```

## ⚖️ Parte 6 — Ética na Apresentação de Dados

No bimestre discutimos a situação do **Jôão**, estagiário que criou gráficos sem seguir as diretrizes do supervisor Carlos.  
As principais regras éticas para visualização de dados são:

1. **Não manipular eixos** — eixos truncados distorcem a percepção de diferenças
2. **Uso neutro de cores** — cores não devem induzir conclusões erradas
3. **Representação clara e precisa** — evitar gráficos 3D ou efeitos distorcivos
4. **Transparência** — indicar claramente médias, medianas ou outras transformações aplicadas

### Exercício reflexivo

Observe os dois gráficos abaixo. Qual é o **ético**? Por quê?


```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico NÃO ético: eixo y truncado
ax1.bar(['A', 'B', 'C'], [8.2, 8.5, 8.8], color='tomato')
ax1.set_ylim(8.0, 9.0)  # eixo truncado — exagera diferenças!
ax1.set_title('❌ Gráfico com eixo truncado\n(Não Ético)', color='red')
ax1.set_ylabel('Nota')

# Gráfico ético: eixo y iniciando em zero
ax2.bar(['A', 'B', 'C'], [8.2, 8.5, 8.8], color='steelblue')
ax2.set_ylim(0, 10)
ax2.set_title('✅ Gráfico com eixo correto\n(Ético)', color='green')
ax2.set_ylabel('Nota')

plt.suptitle('Comparação: Ética na Visualização de Dados', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

Use o DataFrame abaixo com dados de **medalhas** (semelhante ao `dfmedalhas` das aulas) e responda os itens:


```python
# Dataset de medalhas de ouro (EUA, Brasil, China) — 2013 a 2016
dados_medalhas = {
    'Ano':   [2013, 2014, 2015, 2016, 2013, 2014, 2015, 2016, 2013, 2014, 2015, 2016],
    'País':  ['EUA','EUA','EUA','EUA','Brasil','Brasil','Brasil','Brasil','China','China','China','China'],
    'Ouro':  [  45,   50,  48,   46,    12,     15,      11,     14,      38,     42,     40,     44]
}
dfmedalhas = pd.DataFrame(dados_medalhas)
print(dfmedalhas)
```

### Questão 1
Crie um **gráfico de linhas** mostrando a quantidade de medalhas de ouro do **Brasil** ao longo dos anos.


```python
# TODO: escreva seu código aqui
brasil = dfmedalhas[dfmedalhas['País'] == 'Brasil']
ax = brasil.plot(x='Ano', y='Ouro', kind='line', marker='o', figsize=(8, 4))
ax.set_title('Medalhas de Ouro do Brasil (2013–2016)')
ax.set_xlabel('Ano')
ax.set_ylabel('Medalhas de Ouro')
plt.tight_layout()
plt.show()
```

### Questão 2
Crie um **gráfico de barras** com a quantidade total de medalhas de ouro por país (soma de todos os anos).


```python
# TODO: escreva seu código aqui
total_por_pais = dfmedalhas.groupby('País')['Ouro'].sum().reset_index()
ax = total_por_pais.plot(kind='bar', x='País', y='Ouro', color=['green','gold','red'], figsize=(7, 5), legend=False)
ax.set_title('Total de Medalhas de Ouro por País (2013–2016)')
ax.set_xlabel('País')
ax.set_ylabel('Total de Medalhas de Ouro')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Questão 3
Crie um **gráfico de pizza** mostrando a proporção de medalhas de ouro por país.


```python
# TODO: escreva seu código aqui
pizza = total_por_pais.set_index('País')
ax = pizza.plot(kind='pie', y='Ouro', autopct='%1.1f%%', startangle=90,
               colors=['gold','#009C3B','#DE2910'], figsize=(6,6), legend=False)
ax.set_title('Proporção de Medalhas de Ouro por País')
ax.set_ylabel('')
plt.tight_layout()
plt.show()
```

### Questão 4 — Desafio
Use NumPy para calcular a **média** e o **desvio padrão** das medalhas de ouro de cada país separadamente.
Exiba os resultados formatados.


```python
# TODO: complete o código
for pais in dfmedalhas['País'].unique():
    dados = dfmedalhas[dfmedalhas['País'] == pais]['Ouro'].values
    print(f'{pais}: média={np.mean(dados):.2f} | desvio padrão={np.std(dados):.2f}')
```

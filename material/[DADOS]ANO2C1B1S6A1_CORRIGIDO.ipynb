```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.metrics import make_scorer, mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score
import warnings
from sklearn.exceptions import ConvergenceWarning
```

```python
SEED = 42
np.random.seed(SEED)
warnings.filterwarnings("ignore", category=ConvergenceWarning)
```

```python
df = pd.read_csv('/content/house-prices.csv', sep=',')

y = df['Price'].values

df = df.drop('Price', axis=1)

df = pd.get_dummies(df, columns=['Brick', 'Neighborhood'], drop_first=True)

X = df.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED)
```

```python
model_1 = LinearRegression()
```

```python
mae_scorer = make_scorer(mean_squared_error, greater_is_better=False)
mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)
r2_scorer = 'r2'
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED)
```

```python
model_1 = LinearRegression()
```

```python
mae_scorer = make_scorer(mean_squared_error, greater_is_better=False)
mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)
r2_scorer = 'r2'
```

```python
# Scorer para MAE (Mean Absolute Error)
mae_scores = cross_val_score(model_1, X_train, y_train, cv=15, scoring=mae_scorer)
print(f'Cross-Validation MAE Scores: {-mae_scores}')
print(f'Mean MAE Score: {-mae_scores.mean():.4f}')

# Scorer para MSE (Mean Squared Error)
mse_scores = cross_val_score(model_1, X_train, y_train, cv=15, scoring=mse_scorer)
print(f'Cross-Validation MSE Scores: {-mse_scores}')
print(f'Mean MSE Score: {-mse_scores.mean():.4f}')

# RMSE pode ser calculado a partir dos MSE Scores
rmse_scores = np.sqrt(-mse_scores)
print(f'Cross-Validation RMSE Scores: {rmse_scores}')
print(f'Mean RMSE Score: {rmse_scores.mean():.4f}')

# Scorer para R² (Coeficiente de Determinação)
r2_scores = cross_val_score(model_1, X_train, y_train, cv=15, scoring=r2_scorer)
print(f'Cross-Validation R² Scores: {r2_scores}')
print(f'Mean R² Score: {r2_scores.mean():.4f}')
```

```python
# Scorer para MAE (Mean Absolute Error)
mae_scores = cross_val_score(l_2, X_train, y_train, cv=5, scoring=mae_scorer)
print(f'Cross-Validation MAE Scores: {-mae_scores}')
print(f'Mean MAE Score: {-mae_scores.mean():.4f}')

# Scorer para MSE (Mean Squared Error)
mse_scores = cross_val_score(l_2, X_train, y_train, cv=5, scoring=mse_scorer)
print(f'Cross-Validation MSE Scores: {-mse_scores}')
print(f'Mean MSE Score: {-mse_scores.mean():.4f}')

# RMSE pode ser calculado a partir dos MSE Scores
rmse_scores = np.sqrt(-mse_scores)
print(f'Cross-Validation RMSE Scores: {rmse_scores}')
print(f'Mean RMSE Score: {rmse_scores.mean():.4f}')

# Scorer para R² (Coeficiente de Determinação)
r2_scores = cross_val_score(l_2, X_train, y_train, cv=5, scoring=r2_scorer)
print(f'Cross-Validation R² Scores: {r2_scores}')
print(f'Mean R² Score: {r2_scores.mean():.4f}')
```
